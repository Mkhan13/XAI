```python
visited = set()

def cleaning_policy(robot_pos, dirt_map):
    pos = tuple(robot_pos)
    visited.add(pos) # Mark current location as visited

    if dirt_map[pos] == 1: # If current cell is dirty, clean it
      return 4

    neighborPOS = [
        (pos[0]-1, pos[1]), # Left
        (pos[0], pos[1]+1), # Up
        (pos[0]+1, pos[1]), # Right
        (pos[0], pos[1]-1) # Down
    ]

    dirt_spots = [(row, col) for row in range(dirt_map.shape[0]) # Find all dirt locations that have not been visited
                   for col in range(dirt_map.shape[1])
                   if dirt_map[row, col] == 1 and (row, col) not in visited]

    if not dirt_spots: # If there is no more dirt, exit code
        return -1

    distances = [(abs(pos[0]-row) + abs(pos[1]-col), (row, col)) for row, col in dirt_spots] # Find closest dirt block
    _, target = min(distances)

    dRow = target[0] - pos[0] # Vertical direction
    dCol= target[1] - pos[1] # Horizontal direction

    if dRow != 0: # If vertical movement is needed, move up or down
        return 2 if dRow > 0 else 0
    if dCol != 0: # If horizontal movement is needed, move right or left
        return 1 if dCol > 0 else 3

    return -1
```
